```python
def picshow(face_pairs, score=0):
    num_pairs = len(face_pairs)
    fig = plt.figure(figsize=(10, num_pairs * 5))
    for i, match in enumerate(face_pairs):
#         import pdb; pdb.set_trace()
        ax1 = fig.add_subplot(num_pairs, 2, (i * 2) + 1)
        ax2 = fig.add_subplot(num_pairs, 2, (i * 2) + 2)
        ax1.imshow(face_pairs[i].face_1.image)
        ax2.imshow(face_pairs[i].face_2.image)
        ax1.set_title("{} similarity".format(face_pairs[i-1].score))
        ax2.set_title("{} similarity".format(face_pairs[i-1].score))
```

```python
all_faces, all_matches = identifier.find_all_matches('images')
```

```python
picshow(all_matches)
```
